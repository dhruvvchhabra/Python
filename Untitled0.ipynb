```python
students = {
    "Rahul": {"Math": 95, "Science": 88, "English": 76},
    "Anita": {"Math": 67, "Science": 90, "English": 82},
    "Vikram": {"Math": 89, "Science": 92, "English": 91},
    "Sara": {"Math": 78, "Science": 85, "English": 88}
}

# tasks:
# 1. Create a dictionary having each student's average marks
# 2. Find the student with highest average.

average_marks = {}

for student, subjects in students.items():
    total = sum(subjects.values())
    count = len(subjects)
    avg = round(total / count)
    average_marks[student] = avg

print(average_marks)
```
